```python
import matplotlib.pyplot as plt
```

```python
time_intervals = ["12:00 AM", "3:00 AM", "6:00 AM", "9:00 AM", "12:00 PM"]
temperature = [25, 24, 23, 26, 28]
humidity = [60, 65, 70, 75, 80]
```

```python
plt.plot(time_intervals, temperature, marker='o', color='b', label='Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature (Â°C)')
plt.title('Temperature Variation Over Time')
plt.legend()
plt.show()
```

```python
plt.plot(time_intervals, humidity, marker='s', color='g', label='Humidity')
plt.xlabel('Time')
plt.ylabel('Humidity (%)')
plt.title('Humidity Variation Over Time')
plt.legend()
plt.show()
```
