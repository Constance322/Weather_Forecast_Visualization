```python
import pandas as pd

# Load the data from the CSV file into a DataFrame
bottom_countries = pd.read_csv('world_population_data.csv')

# Now you can use the DataFrame and perform operations on it
bottom_countries_sorted = bottom_countries.sort_values(by='growth rate', ascending=False).head(10)['growth rate']
print(bottom_countries_sorted)
```

```python
bottom_countries['growth rate'] = pd.to_numeric(bottom_countries['growth rate'].str.replace('%', ''))

# Now the 'growth rate' column is converted to numeric format without the percentage sign
print(bottom_countries['growth rate'])
```

```python
plt.figure(figsize=(10, 8))
plt.bar(bottom_countries['country'], bottom_countries['growth rate'], color='green')
plt.title("Bottom 10 countries by Growth Rate")
plt.xlabel('Country', rotation=45)  # Corrected parameter name to 'rotation'
plt.ylabel('Growth Rate')
plt.xticks(rotation=60)
plt.show()

```

```python
continent_pop = bottom_countries.groupby('continent')['2023 population'].sum().reset_index()
```

```python
continent_pop
```

```python
# Assuming you have a DataFrame named 'continent_pop' containing the total population for each continent
cp = continent_pop.copy()  # Making a copy of the DataFrame if necessary

plt.figure(figsize=(15, 10))
plt.bar(cp['continent'], cp['2023 population'], color='purple')
plt.title('Population by Continent in 2023')
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.show()

```
